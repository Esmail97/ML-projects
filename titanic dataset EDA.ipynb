#Using URL to load and read dataset

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df1 = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df1
```

#Determine input and output

```python
#drop target column
input = df1.drop('Survived',axis = 'columns')
output = df1.Survived
```

```python
input
```

```python
#convert sex column to numeric values
#using get_dummies
dummies = pd.get_dummies(input.Sex)
dummies
```

```python
#concatenate dummies to the last inputs
input2 = pd.concat([input,dummies],axis = 'columns')
input2.head()
```

```python
#drop the sex and femare or male coulmn
input2.drop(['Sex','male'],axis = 'columns',inplace = True)
```

```python
input2.head()
```

```python
input2.isnull().sum()
```

#fill the null values with the median of Age column

```python
input2['Age'].median()
```

```python
input2.Age = input2.Age.fillna(input2.Age.median())
input2
```
